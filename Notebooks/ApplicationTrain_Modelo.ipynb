### Porcentaje de nulos
Calculemos el porcentaje de nulos de las columnas que no sean mayores al 10%

```python
columnas = [column for column in application_train if application_train[column].isnull().mean()>0.1]
application_train = application_train.drop(columnas,axis=1)
df_null = pd.DataFrame(application_train.isnull().mean())
df_null.rename(columns={0:'% nulos'},inplace=True)
df_null = df_null[df_null['% nulos']!=0]
df_null.sort_values('% nulos',ascending=False,inplace=True)
df_null
```

podemos ver que la cantidad de nulos de las columnas que no son mayores al 10% son muy bajos, que pueden ser quitados directamente de los datos
